```python
text='Happy hour12-3'
l=''
for i in range(len(text)):
    if text [i].isupper():
        l=l+text[i].lower()
    elif text[i].islower():
        l=l+text[i].upper()
    elif text[i].isdigit():
        l=l+(text[i])
    else:
        l=l+'#'
print(l)
```
